```python
#Funcion para ver las graficas de los 4 miembros de cada uno de los pacientes:
def Paciente(FileName, FigsizeX, FigsizeY):
    Valores = []
    datosN = pd.read_csv(FileName)
    datosN.index = ['MSI','MSD','MII','MID']
    #datosN = datosN.drop(['0.1Hz'], axis=0)
    DN = datosN.to_numpy().tolist()
    #Debemos recordar sumar 1 al final del arreglo por la cuenta de python
    freqs = [0.1, 0.2, 0.3, 0.5, 1, 2, 5] 
    MSIN = DN[0]
    MSDN = DN[1]
    MIIN = DN[2]
    MIDN = DN[3]
    Valores.append(MSIN)
    Valores.append(MSDN)
    Valores.append(MIIN)
    Valores.append(MIDN)
    plt.figure(figsize=(FigsizeX, FigsizeY))
    plt.subplot(2,2,1)
    plt.scatter(freqs, MSIN, color = 'green')
    plt.xlabel('Frecuencia')
    plt.ylabel('%DDF')
    plt.ylim(-100,100)
    plt.xlim(0,5.5)
    plt.title('Miembro Sup. Izq.')

    plt.subplot(2,2,2)
    plt.scatter(freqs, MSDN, color = 'orange')
    plt.xlabel('Frecuencia')
    #plt.ylabel('%DDF')
    plt.ylim(-100,100)
    plt.xlim(0,5.5)
    plt.title('Miembro Sup. Der.')
    plt.figure(2)
    
    plt.figure(figsize=(FigsizeX, FigsizeY))
    plt.subplot(2,2,3)
    plt.scatter(freqs, MIIN, color = 'red')
    plt.xlabel('Frecuencia')
    plt.ylabel('%DDF')
    plt.ylim(-100,100)
    plt.xlim(0,5.5)
    plt.title('Miembro Inf. Izq.')

    plt.subplot(2,2,4)
    plt.scatter(freqs, MIDN, color = 'blue')
    plt.title('Miembro Inf. Der.')
    plt.xlabel('Frecuencia')
    #plt.ylabel('%DDF')
    plt.ylim(-100,100)
    plt.xlim(0,5.5)
    plt.show()
    
    return Valores
```

```python
# Funcion para los promedios de todos los miembros para todos los pacientes:
#n:=Numero de filas totales, en este caso para 8 pacientes multiplicado por los 4 miembros n=8x4=32
def Promedios(NombreTabla, n):
    lista_promedios=[]
    promedios_MSI, promedios_MSD, promedios_MII, promedios_MID=[], [], [], []
    S_MSI_01, S_MSI_02, S_MSI_03, S_MSI_05, S_MSI_1, S_MSI_2, S_MSI_5 = 0, 0, 0, 0, 0, 0, 0
    S_MSD_01, S_MSD_02, S_MSD_03, S_MSD_05, S_MSD_1, S_MSD_2, S_MSD_5 = 0, 0, 0, 0, 0, 0, 0
    S_MII_01, S_MII_02, S_MII_03, S_MII_05, S_MII_1, S_MII_2, S_MII_5 = 0, 0, 0, 0, 0, 0, 0
    S_MID_01, S_MID_02, S_MID_03, S_MID_05, S_MID_1, S_MID_2, S_MID_5 = 0, 0, 0, 0, 0, 0, 0 
    a=0
    for i in range(1, n+1): #entramos a la suma
        if a==0: #a==0 para que slo lo haga una vez (porque se actualiza esa variable).
            for u in range(0,n*4,4): #Para 0.1Hz donde la columna es 0:
                S_MSI_01 = S_MSI_01 + NombreTabla[0][u]
            h = S_MSI_01/n
            promedios_MSI.append(h) #Lo guardo
            for u in range(0,n*4,4): #Para 0.2Hz donde la columna es 1:
                S_MSI_02 = S_MSI_02 + NombreTabla[1][u]
            b = S_MSI_02/n
            promedios_MSI.append(b)
            for u in range(0,n*4,4): #Para 0.3Hz donde la columna es 2:
                S_MSI_03 = S_MSI_03 + NombreTabla[2][u]
            c = S_MSI_03/n
            promedios_MSI.append(c)
            for u in range(0,n*4,4): #Para 0.5Hz donde la columna es 3:
                S_MSI_05 = S_MSI_05 + NombreTabla[3][u]
            d = S_MSI_05/n
            promedios_MSI.append(d)
            for u in range(0,n*4,4): #Para 1Hz donde la columna es 4:
                S_MSI_1 = S_MSI_1 + NombreTabla[4][u]
            e = S_MSI_1/n
            promedios_MSI.append(e)
            for u in range(0,n*4,4): #Para 2Hz donde la columna es 5:
                S_MSI_2 = S_MSI_2 + NombreTabla[5][u]
            f = S_MSI_2/n
            promedios_MSI.append(f)
            for u in range(0,n*4,4): #Para 5Hz donde la columna es 6:
                S_MSI_5 = S_MSI_5 + NombreTabla[6][u]
            g = S_MSI_5/n
            promedios_MSI.append(g)
#################################################################################################Miembro MSD 7 freqs:
            for u in range(1,n*4,4): #Para 0.1Hz donde la columna es 0:
                S_MSD_01 = S_MSD_01 + NombreTabla[0][u]
            hh = S_MSD_01/n
            promedios_MSD.append(hh)
            for u in range(1,n*4,4): #Para 0.2Hz donde la columna es 1:
                S_MSD_02 = S_MSD_02 + NombreTabla[1][u]
            bb = S_MSD_02/n
            promedios_MSD.append(bb)
            for u in range(1,n*4,4): #Para 0.3Hz donde la columna es 2:
                S_MSD_03 = S_MSD_03 + NombreTabla[2][u]
            cc = S_MSD_03/n
            promedios_MSD.append(cc)
            for u in range(1,n*4,4): #Para 0.5Hz donde la columna es 3:
                S_MSD_05 = S_MSD_05 + NombreTabla[3][u]
            dd = S_MSD_05/n
            promedios_MSD.append(dd)
            for u in range(1,n*4,4): #Para 1Hz donde la columna es 4:
                S_MSD_1 = S_MSD_1 + NombreTabla[4][u]
            ee = S_MSD_1/n
            promedios_MSD.append(ee)
            for u in range(1,n*4,4): #Para 2Hz donde la columna es 5:
                S_MSD_2 = S_MSD_2 + NombreTabla[5][u]
            ff = S_MSD_2/n
            promedios_MSD.append(ff)
            for u in range(1,n*4,4): #Para 5Hz donde la columna es 6:
                S_MSD_5 = S_MSD_5 + NombreTabla[6][u]
            gg = S_MSD_5/n
            promedios_MSD.append(gg)
###############################################################################################Miembro MII 7 freqs:
            for u in range(2,n*4,4): #Para 0.1Hz donde la columna es 0:
                S_MII_01 = S_MII_01 + NombreTabla[0][u]
            hhh = S_MII_01/n
            promedios_MII.append(hhh)
            for u in range(2,n*4,4): #Para 0.2Hz donde la columna es 1:
                S_MII_02 = S_MII_02 + NombreTabla[1][u]
            bbb = S_MII_02/n
            promedios_MII.append(bbb)
            for u in range(2,n*4,4): #Para 0.3Hz donde la columna es 2:
                S_MII_03 = S_MII_03 + NombreTabla[2][u]
            ccc = S_MII_03/n
            promedios_MII.append(ccc)
            for u in range(2,n*4,4): #Para 0.5Hz donde la columna es 3:
                S_MII_05 = S_MII_05 + NombreTabla[3][u]
            ddd = S_MII_05/n
            promedios_MII.append(ddd)
            for u in range(2,n*4,4): #Para 1Hz donde la columna es 4:
                S_MII_1 = S_MII_1 + NombreTabla[4][u]
            eee = S_MII_1/n
            promedios_MII.append(eee)
            for u in range(2,n*4,4): #Para 2Hz donde la columna es 5:
                S_MII_2 = S_MII_2 + NombreTabla[5][u]
            fff = S_MII_2/n
            promedios_MII.append(fff)
            for u in range(2,n*4,4): #Para 5Hz donde la columna es 6:
                S_MII_5 = S_MII_5 + NombreTabla[6][u]
            ggg = S_MII_5/n
            promedios_MII.append(ggg)
##############################################################################################Miembro MID 7 freqs:
            for u in range(3,n*4,4): #Para 0.1Hz donde la columna es 0:
                S_MID_01 = S_MID_01 + NombreTabla[0][u]
            hhhh = S_MID_01/n
            promedios_MID.append(hhhh)
            for u in range(3,n*4,4): #Para 0.2Hz donde la columna es 1:
                S_MID_02 = S_MID_02 + NombreTabla[1][u]
            bbbb = S_MID_02/n
            promedios_MID.append(bbbb)
            for u in range(3,n*4,4): #Para 0.3Hz donde la columna es 2:
                S_MID_03 = S_MID_03 + NombreTabla[2][u]
            cccc = S_MID_03/n
            promedios_MID.append(cccc)
            for u in range(3,n*4,4): #Para 0.5Hz donde la columna es 3:
                S_MID_05 = S_MID_05 + NombreTabla[3][u]
            dddd = S_MID_05/n
            promedios_MID.append(dddd)
            for u in range(3,n*4,4): #Para 1Hz donde la columna es 4:
                S_MID_1 = S_MID_1 + NombreTabla[4][u]
            eeee = S_MID_1/n
            promedios_MID.append(eeee)
            for u in range(3,n*4,4): #Para 2Hz donde la columna es 5:
                S_MID_2 = S_MID_2 + NombreTabla[5][u]
            ffff = S_MID_2/n
            promedios_MID.append(ffff)
            for u in range(3,n*4,4): #Para 5Hz donde la columna es 6:
                S_MID_5 = S_MID_5 + NombreTabla[6][u]
            gggg = S_MID_5/n
            promedios_MID.append(gggg)
            a=a+1 #actualizo la variable para que el ciclo no se repita
    
    lista_promedios.append(promedios_MSI)
    lista_promedios.append(promedios_MSD)
    lista_promedios.append(promedios_MII)
    lista_promedios.append(promedios_MID)
    return lista_promedios
```
